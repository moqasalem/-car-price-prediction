Step 1 load data

```python
import pandas as pd
df = pd.read_csv("CarPrice_Assignment.csv", header=0, sep =",")
print(df.head())
```

```python
df.isnull().sum()
```

```python
df.info()
```

Step 2 List of categorical columns

```python
#Drop ID because not feature
df.drop('car_ID', axis=1, inplace=True)

# List of categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

# Create dummy variables
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)

print(df_encoded.shape)
```

Step 4 Correlation Matrix

```python
corr_matrix = df_encoded.corr()

#Price important var 
corr_with_price = corr_matrix["price"].sort_values(ascending=False)
print("Correlation with price:\n", corr_with_price)
```

```python
#heatmap 
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
sns.heatmap(df_encoded.corr(), cmap="coolwarm")
```

```python
#draw Correlations

#enginesize vs price
import matplotlib.pyplot as plt
import numpy as np

x = df['enginesize']
y = df['price']

#  scatter
plt.scatter(x, y, color='blue', label='Data Points')

#     (degree=1   )
m, b = np.polyfit(x, y, 1)   # m = slope, b = intercept

#   
plt.plot(x, m*x + b, color='red', label='Trend Line')

plt.xlabel("Engine Size")
plt.ylabel("Price")
plt.title("Engine Size vs Price with Trend Line")
plt.legend()
plt.show()
```

```python
#example to predict: 
enginesize_new = 260
predicted_price = m * enginesize_new + b
print("Predicted Price:", predicted_price)
```
